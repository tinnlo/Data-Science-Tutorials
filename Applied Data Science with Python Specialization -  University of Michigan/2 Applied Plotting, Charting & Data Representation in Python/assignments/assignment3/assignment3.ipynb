In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])

df
```

```python
%matplotlib widget

from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.cm as cm

####### 1. Prepare the data for the plot #######
# Get the means and standard deviations
df_mean = df.mean(axis=1)
df_std = df.std(axis=1)

# the number of the records of a year
n = df.shape[1]

# Set y default value as the mean of 4 bars / all data
y = np.mean(df_mean.values)

# calculates the Z-score for a 95% confidence interval (two-tailed)
yerr = df_std / np.sqrt(n) * stats.norm.ppf(1 - 0.05 / 2) 

# compute the 95% confidence intervals
conf_ints=[]
for idx in df.index:
    conf_int = stats.norm.interval(0.95, np.mean(df.loc[idx]), stats.sem(df.loc[idx])) # 95%, mean, se
    conf_ints.append(conf_int)

# A function for computing the probabilities
def compute_probs(y, conf_int):
    if y < np.min(conf_int):
        result = 1.0
    elif y > np.max(conf_int):
        result = 0.0
    else:
        result = (np.max(conf_int) - y) / (np.max(conf_int) - np.min(conf_int))
    return result

# call the function to generate probabilities series
probs = [compute_probs(y, ci) for ci in conf_ints]


####### 2. Setup the plot #######
plt.figure(figsize=(8, 8))

# Setup the colormap
cmap = cm.get_cmap('coolwarm')
sm = cm.ScalarMappable(cmap = cmap, norm = plt.Normalize(vmin=0, vmax=1))
sm.set_array([])

# Setup the bars
bars = plt.bar(range(len(df)),
               df_mean,
               yerr=yerr,
               alpha=0.8,
               color=sm.to_rgba(probs), #interpret the probabilities to colors
               capsize=7)


# Add the colorbar lengend
cbar = plt.colorbar(sm, orientation = "horizontal")

# Turn off some plot rectangle spines
[plt.gca().spines[loc].set_visible(False) for loc in ['top', 'right']]

# Set ticks and labels
plt.title('Even Harder: interact with y axis', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Value')
plt.xticks(range(len(df)), df.index)

# Add the probabilities of each bar
text1=plt.text(bars[0].get_x()+bars[0].get_width()/2,bars[0].get_height()+9500,'prob={:.2f}'.format(1-abs(probs[0])),bbox={'fc':'w','ec':'k'},ha='center')
text2=plt.text(bars[1].get_x()+bars[1].get_width()/2,bars[1].get_height()+6000,'prob={:.2f}'.format(1-abs(probs[1])),bbox={'fc':'w','ec':'k'},ha='center')
text3=plt.text(bars[2].get_x()+bars[2].get_width()/2,bars[2].get_height()+7500,'prob={:.2f}'.format(1-abs(probs[2])),bbox={'fc':'w','ec':'k'},ha='center')
text4=plt.text(bars[3].get_x()+bars[3].get_width()/2,bars[3].get_height()+5000,'prob={:.2f}'.format(1-abs(probs[3])),bbox={'fc':'w','ec':'k'},ha='center')


####### 3. The interactive y axis #######
# set the horizontal line
hoz_line = plt.axhline(y=y, color='gray', linewidth=1, linestyle='--')
# get the existed y ticks
yticks = plt.gca().get_yticks()
# add the interactive line to the y ticks
new_yticks=np.append(yticks,y)
# generate new y ticks with the interactive y axis
plt.gca().set_yticks(new_yticks)

# set the annotation for the interactive y axis
# set the text box with face color (fc) to white ('w') 
# and the edge color (ec) to black ('k')
y_text = plt.text(1.2, 55000, 'y = %d' % y, bbox=dict(fc='w', ec='k')) 

# Add interactivity with onclick event
def onclick(event):

    # get the new y axis position
    y = event.ydata
    # change the y line
    hoz_line.set_ydata(event.ydata)
    # get the new y tick value
    new_yticks=np.append(yticks,y)
    # update the new y tick on the plot
    plt.gca().set_yticks(new_yticks)
    
    # compute the probabilites
    probs = [compute_probs(y, ci) for ci in conf_ints]
    
    # Add the probabilities of each bar
    y_text.set_text('y={:5.0f}'.format(event.ydata))
    text1.set_text('prob={:.2f}'.format(abs(probs[0])))
    text2.set_text('prob={:.2f}'.format(abs(probs[1])))
    text3.set_text('prob={:.2f}'.format(abs(probs[2])))
    text4.set_text('prob={:.2f}'.format(abs(probs[3]))) 
    
    #print(probs)
    for i in range(len(df)):
        bars[i].set_color(sm.to_rgba(probs[i]))
    
plt.gcf().canvas.mpl_connect('button_press_event', onclick)

####### 4. Display the plot #######
plt.show()
```
